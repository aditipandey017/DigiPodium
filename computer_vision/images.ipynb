```python
avengers= cv2.imread('avengers.jpg')    #inverted colors becoz it read data in BGR instead of RGB
```

```python
plt.imshow(avengers)
plt.axis('off')
plt.show()
```

```python
avg_converted= cv2.cvtColor(avengers,cv2.COLOR_BGR2RGB)
plt.imshow(avg_converted)
plt.axis('off')
```

```python
avg_gray= cv2.cvtColor(avengers,cv2.COLOR_BGR2GRAY)
plt.imshow(avg_gray, cmap='gray')
plt.axis('off')
```

```python
plt.imshow(np.hstack([avengers,avg_converted]))
plt.axis('off')
```
